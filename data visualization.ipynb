```python
import matplotlib.pyplot as plt
import numpy as np

style.use('ggplot')

# Create a list of time hours and number of users
time_hrs = [4, 3, 5, 6, 9, 10, 11, 12, 13, 15, 17]
web_cust = [345, 1359, 353, 66, 854, 554, 846, 653, 12, 433]

# Add a dummy element to the web_cust list
web_cust.append(0)

# Plot the number of users against the time of day
plt.plot(time_hrs, web_cust)

# Set the axis labels and title
plt.xlabel('Hrs')
plt.ylabel('NO.OF Users')
plt.title('Web site traffic')

# Show the plot
plt.show()
```

controlling line patterns and colors

```python
plt.plot(time_hrs, web_cust,color='g',linestyle='--',linewidth=2)

# Set the axis labels and title
plt.xlabel('Hrs')
plt.ylabel('NO.OF Users')
plt.title('Web site traffic')

# Show the plot
plt.show()
```

set axis, labels, & Legend property

```python
plt.plot(time_hrs, web_cust,'g',label='web traffic',linewidth=2)
#plt.axis(6.5,17.5,50,2000)

# Set the axis labels and title
plt.xlabel('Hrs')
plt.ylabel('NO.OF Users')
plt.title('Web site traffic')
plt.legend()
# Show the plot
plt.show()
```

Alpha and Annotation

Annotate() mthd is used to annotate the graph.It has several attributes which helps to annotate the plot.

"MAX" denotes the annotation text.

"ha" indicates the horziontal alignment

"va" indicates vertical alignment 

"xytext" specifies the position of the text in data coordinates

"xy" indicates the arrow position

"arrowprops" are properties for drawing arrow

```python
style.use('ggplot')
plt.plot(time_hrs,web_cust,alpha=.8)
plt.title('Website Traffic')
plt.annotate('Max',ha='center',va='bottom',xytext=(8,1500),xy=(11,1630),arrowprops={
    'facecolor':'blue'})
# Set the axis labels and title
plt.xlabel('Hrs')
plt.ylabel('NO.OF Users')

# Show the plot
plt.show()
```

## Multiple Plots

```python
style.use('ggplot')

web_mon = [345, 1359, 353, 66, 854, 554, 846, 653, 12, 433]
web_tues=[56,12,58,123,66,45,57,9,99,102,47]
web_wed=[88,45,53,39,68,77,23,91,54,63,35]
time_hrs = [4, 3, 5, 6, 9, 10, 11, 12, 13, 15, 17]

# Add a dummy element to the web_mon list
web_mon.append(0)

plt.plot(time_hrs,web_mon,'g',label='Monday',linewidth=4)
plt.plot(time_hrs,web_tues,'r',label='Tuesday',linewidth=3)
plt.plot(time_hrs,web_wed,'b',label='Wednesday',linewidth=2)
plt.axis([6.5,17.5,50,200])
plt.title('Website Traffic using multiple plots')
plt.xlabel('Hrs')
plt.ylabel('NO.OF Users')
plt.legend()
# Show the plot
plt.show()
```

# Subplots 
subplot is used to display multiple plots in the same window, arrange plots in a regular grid

Layout and spacing Adjustments are two important factors to be considered while creating subplots.

use plt.subplots_adjust() mthd with the parameters hspace and wspace to adust the distance between the subplot and move them around on the grid.


```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
temp_data=[79,84,35,54,66,65,35,69,23,42,5]
wind_data=[88,45,53,39,68,77,23,91,54,63,35]
time_hrs=[78,55,66,12,89,79,99,26,97,55,95]
humidity_data=[76,99,56,22,41,44,43,92,25,47,9]
precep_data=[80,44,67,89,33,9,77,63,72,87,35]
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Temp')
plt.plot(time_hrs,temp_data,color='b',linestyle='-',linewidth=2)
plt.subplot(1,2,2)
plt.title('Wind')
plt.plot(time_hrs,wind_data,color='g',linestyle="-",linewidth=1)
```

```python
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,1,1)
plt.title('Humidity')
plt.plot(time_hrs,humidity_data,color='b',linestyle='-',linewidth=1)
plt.title('precipitation')
plt.plot(time_hrs,precep_data,color='g',linestyle='-',linewidth=2)
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.subplots_adjust(hspace=.3)
plt.subplot(2,2,1)
plt.title('Temp (F)')
plt.plot(time_hrs,temp_data,color='b',linestyle='-',linewidth=1)
plt.subplot(2,2,2)
plt.title('Wind (mph)')
plt.plot(time_hrs,wind_data,color='y',linestyle='-',linewidth=2)
plt.subplot(2,2,3)
plt.title('Humidity (%)')
plt.plot(time_hrs,humidity_data,color='y',linewidth=3,linestyle='-')
plt.subplot(2,2,4)
plt.title('precipitation')
plt.plot(time_hrs,precep_data,color='g',linestyle='-',linewidth=4)
plt.show()
```
